```python
def factorial(n):
    if n == 0 :
        return 1
    else :
        return n* factorial(n-1)
print(factorial(5))
```

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    
    for char in text:
        if char in vowels:
            count += 1
            
    return count

# Check karne ke liye example:
print(count_vowels("Hello World"))  # Output: 3
```

```python
def find_maximum(numbers_list):
    if not numbers_list:
        return None  # Agar list khali hai toh None return karega
    return max(numbers_list)

# Test karna:
my_list = [10, 20, 30, 70, 40]
print(find_maximum(my_list))  # Output: 70
```

```python
def is_palindrome(text):
    # Saree characters ko lowercase karke reverse string se compare karte hain
    clean_text = text.lower()
    return clean_text == clean_text[::-1]

# Test karna:
print(is_palindrome("radar"))  # Output: True
print(is_palindrome("hello"))  # Output: False
```

```python
def sum_up_to_n(n):
    # Mathematical formula ka use karke integer value return karega
    return (n * (n + 1)) // 2

# Test karna:
print(sum_up_to_n(5))   
print(sum_up_to_n(10))  
```
