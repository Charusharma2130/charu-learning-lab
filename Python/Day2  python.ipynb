# Python compound interest calculater

```python
principle = 0
rate = 0
time = 0
principle = int(input("enter the principle amount:"))
    
while principle <= 0 :
    print("principle are not negative ")
    principle = int(input("enter the principle amount:"))

rate = int(input("enter the rate:"))
while rate <= 0 :
    print("rate are not negative ")
    rate = int(input("enter the rate :"))

time = int(input("enter the time :"))
while time <= 0 :
    print("time  are not negative ")
    time = int(input("enter the time :"))
Total = principle*((1+rate/100)**time)
print(f"your compound interest is ${Total}")
  
```

```python
import time
my_time = int(input("enter is time in second"))
for X in range(my_time,0,-1):
    second = X%60
    minute = int(X/60)%60
    hour = int(X/3600)
    print(f"{hour:02}:{minute:02}:{second:02}")
    time.sleep(1)
print("TIME'S UP")    
```
