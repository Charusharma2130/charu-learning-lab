```python
#Check karo string palindrome hai ya nahi.

def is_palindrome(text):
    # Saree characters ko lowercase karke reverse string se compare karte hain
    clean_text = text.lower()
    return clean_text == clean_text[::-1]

# Test karna:
print(is_palindrome("radar"))  
print(is_palindrome("hello"))  
```

```python
def find_max_char(my_string):
    # 1. एक खाली डिक्शनरी बनाएं गिनती रखने के लिए
    count_dict = {}
    
    # 2. स्ट्रिंग के हर कैरेक्टर पर लूप चलाएं
    for char in my_string:
        if char in count_dict:
            count_dict[char] += 1  # अगर कैरेक्टर पहले से है, तो गिनती +1 करें
        else:
            count_dict[char] = 1   # अगर पहली बार आया है, तो गिनती 1 सेट करें
            
    # 3. सबसे ज़्यादा बार आने वाला कैरेक्टर ढूंढें
    max_char = ""
    max_count = 0
    
    for char in count_dict:
        if count_dict[char] > max_count:
            max_count = count_dict[char]
            max_char = char
            
    return max_char, max_count

# टेस्ट करने के लिए
text = "apple"
char, count = find_max_char(text)
print(f"सबसे ज़्यादा आने वाला अक्षर '{char}' है, जो {count} बार आया है।")
```
